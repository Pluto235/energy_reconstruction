# Core Selection Compare 2724

这个 notebook 用来对比 Slurm job `2724` 的三个 core selection 实验结果：

1. `corebox_true_2724`: true-core selection (`use_core_box=true`, `dcedge_min=0`)
2. `dcedge20_2724`: reconstructed core cut at 20 m (`dcedge_min=20`)
3. `no_core_cut_2724`: no core selection (`use_core_box=false`, `dcedge_min=0`)

数据自动读取自：

- `runs/*/fig/preds.npz`
- `runs/*/fig/metrics.json`
- `runs/*/config.json`

下方三张对比图会复用 `preds.npz` 中的 `logE_true`、`logE_pred`、`mc_weight`，并按照 `src/common/utils.py` 中相同的加权统计口径，重新计算并叠加三条曲线：

- `resolution_weighted`
- `logRMS_weighted`
- `bias_weighted`


## 1. 设置路径并导入当前项目代码

这一格只做环境准备：定位项目根目录、输出目录，并导入当前整理后的 `src/common` 工具函数。

```python
from pathlib import Path
import json
import sys

import numpy as np
import matplotlib.pyplot as plt

PROJECT_ROOT = Path('/home/server/projects/energy_reconstruction')
NOTEBOOK_DIR = PROJECT_ROOT / 'notebook'
OUTPUT_DIR = NOTEBOOK_DIR / 'generated' / 'core_selection_compare_2724'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.common.utils import _weighted_mean, _weighted_std, _weighted_rms

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

print('PROJECT_ROOT =', PROJECT_ROOT)
print('OUTPUT_DIR   =', OUTPUT_DIR)
```

## 2. 定义实验目录并检查结果文件

这里把 job `2724` 的三个 run 明确绑定起来，并确认 notebook 后续要读取的三个文件都存在。

```python
RUNS = {
    'corebox_true': {
        'run_name': 'corebox_true_2724',
        'display': 'True core selection',
    },
    'dcedge20': {
        'run_name': 'dcedge20_2724',
        'display': 'Reconstructed core cut (20 m)',
    },
    'no_core_cut': {
        'run_name': 'no_core_cut_2724',
        'display': 'No core selection',
    },
}

RESULTS = {}
for key, item in RUNS.items():
    run_dir = PROJECT_ROOT / 'runs' / item['run_name']
    fig_dir = run_dir / 'fig'
    config_path = run_dir / 'config.json'
    metrics_path = fig_dir / 'metrics.json'
    preds_path = fig_dir / 'preds.npz'

    assert run_dir.exists(), f'Missing run dir: {run_dir}'
    assert config_path.exists(), f'Missing config: {config_path}'
    assert metrics_path.exists(), f'Missing metrics: {metrics_path}'
    assert preds_path.exists(), f'Missing preds: {preds_path}'

    RESULTS[key] = {
        'display': item['display'],
        'run_dir': run_dir,
        'config_path': config_path,
        'metrics_path': metrics_path,
        'preds_path': preds_path,
    }

for key, item in RESULTS.items():
    print(f"{key:12s} -> {item['run_dir']}")
    print(f"  preds   : {item['preds_path']}")
    print(f"  metrics : {item['metrics_path']}")
    print(f"  config  : {item['config_path']}")
```

## 3. 读取 `config.json`、`metrics.json` 和 `preds.npz`

这一步不手工抄数据，直接从结果文件自动读取。`preds.npz` 会提供重新画曲线所需的原始数组。

```python
loaded = {}
summary_rows = []

for key, item in RESULTS.items():
    with open(item['config_path'], 'r') as f:
        config = json.load(f)
    with open(item['metrics_path'], 'r') as f:
        metrics = json.load(f)
    preds = np.load(item['preds_path'])

    loaded[key] = {
        'display': item['display'],
        'config': config,
        'metrics': metrics,
        'logE_true': preds['logE_true'].astype(np.float64),
        'logE_pred': preds['logE_pred'].astype(np.float64),
        'mc_weight': preds['mc_weight'].astype(np.float64),
    }

    summary_rows.append({
        'experiment': item['display'],
        'run_name': item['run_dir'].name,
        'n': metrics['n'],
        'w_log_sigma': metrics['w_log_sigma'],
        'w_log_rmse': metrics['w_log_rmse'],
        'w_log_bias': metrics['w_log_bias'],
    })

for row in summary_rows:
    print(row)
```

## 4. 用共享 true-energy bins 重算三条加权曲线

这里用 `src/common.utils` 中相同的加权统计定义：

- `bias_weighted`: 每个 true-energy bin 内的加权平均残差 `logE_pred - logE_true`
- `resolution_weighted`: 每个 true-energy bin 内的加权 `std(logE_pred)`
- `logRMS_weighted`: 每个 true-energy bin 内的加权 RMS 残差

为了让三组实验可以直接叠图对比，这里使用三组样本共同的 true-energy 范围，构造统一的 bin edges。

```python
def compute_weighted_curves(log_true, log_pred, weights, bin_edges):
    mask = (
        np.isfinite(log_true)
        & np.isfinite(log_pred)
        & np.isfinite(weights)
        & (weights > 0)
    )

    log_true = log_true[mask]
    log_pred = log_pred[mask]
    weights = weights[mask]

    centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
    bias = []
    resolution = []
    log_rms = []

    for i in range(len(bin_edges) - 1):
        m = (log_true >= bin_edges[i]) & (log_true < bin_edges[i + 1])
        if m.sum() > 10:
            residual = log_pred[m] - log_true[m]
            w = weights[m]
            bias.append(float(_weighted_mean(residual, w)))
            resolution.append(float(_weighted_std(log_pred[m], w)))
            log_rms.append(float(_weighted_rms(residual, w)))
        else:
            bias.append(np.nan)
            resolution.append(np.nan)
            log_rms.append(np.nan)

    return {
        'bin_centers': centers,
        'bias': np.asarray(bias, dtype=np.float64),
        'resolution': np.asarray(resolution, dtype=np.float64),
        'log_rms': np.asarray(log_rms, dtype=np.float64),
    }

all_true = []
for item in loaded.values():
    mask = np.isfinite(item['logE_true'])
    all_true.append(item['logE_true'][mask])

combined_true = np.concatenate(all_true)
bin_edges = np.linspace(combined_true.min(), combined_true.max(), 21)

curves = {}
for key, item in loaded.items():
    curves[key] = compute_weighted_curves(
        item['logE_true'],
        item['logE_pred'],
        item['mc_weight'],
        bin_edges,
    )
    valid_bins = np.isfinite(curves[key]['bias']).sum()
    print(f"{item['display']}: valid bins = {valid_bins}")
```

## 5. 定义统一绘图函数

接下来三张图都使用同一套颜色、线型和图例风格，方便直接比较。

```python
plot_order = ['corebox_true', 'dcedge20', 'no_core_cut']
colors = {
    'corebox_true': '#1f77b4',
    'dcedge20': '#ff7f0e',
    'no_core_cut': '#2ca02c',
}
markers = {
    'corebox_true': 'o',
    'dcedge20': 's',
    'no_core_cut': '^',
}

def plot_compare(metric_key, ylabel, title, filename, draw_zero=False):
    fig, ax = plt.subplots(figsize=(8, 5))
    for key in plot_order:
        item = loaded[key]
        curve = curves[key]
        ax.plot(
            curve['bin_centers'],
            curve[metric_key],
            marker=markers[key],
            color=colors[key],
            linewidth=2,
            markersize=5,
            label=item['display'],
        )

    if draw_zero:
        ax.axhline(0.0, color='gray', linestyle='--', linewidth=1)

    ax.set_xlabel(r'True Energy $\log_{10}(E/\mathrm{GeV})$')
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend(frameon=True)
    ax.grid(alpha=0.3)
    fig.tight_layout()

    out_path = OUTPUT_DIR / filename
    fig.savefig(out_path, dpi=200)
    plt.show()
    print('saved:', out_path)

    return out_path
```

## 6. 画 `resolution_weighted` 对比图

同一张图中叠加三组实验的 weighted resolution 曲线。

```python
resolution_path = plot_compare(
    metric_key='resolution',
    ylabel='Weighted resolution',
    title='Core selection comparison: weighted resolution',
    filename='resolution_weighted_compare_2724.png',
)
resolution_path
```

## 7. 画 `logRMS_weighted` 对比图

同一张图中叠加三组实验的 weighted log RMS 曲线。

```python
logrms_path = plot_compare(
    metric_key='log_rms',
    ylabel='Weighted log RMS error',
    title='Core selection comparison: weighted log RMS',
    filename='logRMS_weighted_compare_2724.png',
)
logrms_path
```

## 8. 画 `bias_weighted` 对比图

同一张图中叠加三组实验的 weighted bias 曲线，并画出零偏差参考线。

```python
bias_path = plot_compare(
    metric_key='bias',
    ylabel='Weighted bias',
    title='Core selection comparison: weighted bias',
    filename='bias_weighted_compare_2724.png',
    draw_zero=True,
)
bias_path
```
